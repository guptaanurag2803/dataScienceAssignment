```python
#KeyError

# KeyError exception is raised when a key is not found in a dictionary or other mapping object.

dic = {'abc': 1, 'xyz': 2}

# In the above dictionary if we try to access element with key pqr (dic['pqr']), it will throw an error.
# This is because there no element exists with key pqr.
```

```python
#IndexError

# IndexError exception is raised when an index is out of range in a sequence such as a list or a string.

s = "Hello"

# In above string if we try to access 7th element from string s (s[6]),  it will generate an error.
# This is because there is no 7th element exists in string s.
```
