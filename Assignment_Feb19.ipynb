```python
# 1.

# Flask is a lightweight web application framework written in Python.
# Flask allows developers to quickly build web applications with minimum overhead.

# ADVANTAGES:
# Lightweight
# Built-in development server
# Easy to learn
# More flexible
# RESTful API support
```

```python
#2.

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#3.

# App rounting is used to map a URL with a specific python function which is intended to perform a specific task.

# App routes make it easier to remember the URLs and also helps in making the navigation easier.
```

```python
#4.

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999</h1>"

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation<h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<int:user_id>')
def show_user(user_id):
    return f'This is user {user_id}'

if __name__=="__main__":
    app.run(host="0.0.0.0")
    
```
