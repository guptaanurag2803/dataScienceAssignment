```python
#1. 

# GET method is used by the client to retrieve data from the server.
# When a client makes a GET request, it sends a URL and any parameters as part of the URL.

# POST method is used to submit data to the server.
# When a client makes a POST request, it sends data in the request body, typically in the form of a form submission or JSON data.
```

```python
#2.

# 'request' is an object which contains all the data sent by the client to the server.
# We can use GET or POST methods to get this data.
# With the help of request, we can easily extract information from incoming requests and use it to generate dynamic responses.
```

```python
#3.

# The redirect() function in Flask is used to redirect the user to a different URL or endpoint.
# It is commonly used in web applications to redirect the user to a new page after they have submitted a form or taken some other action.
```

```python
#4.

# templates are used to generate dynamic HTML pages as the data or input is given.

# The render_template() function is used to render a template and return the resulting HTML to the client.
# The function takes a template name as its first argument and any additional arguments are passed to the template context, which allows you to pass dynamic data to the template.
```

```python
#5.

from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/', methods=['POST'])
def home():
    if (request.method == "POST"):
        name = request.json['name']
        marks = int(request.json['marks'])
        result = f"Name : {name} and Marks : {marks}"
    return jsonify(result)
            
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
