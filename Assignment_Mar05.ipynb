```python
#1.

# First, we have to import bokeh.io and bokeh.plotting
# Now, we have to use bokeh.io.output_notebook()
# Now import figure, output_file and show from bokeh.plotting
# Use test.html as output file and give title, x axis label and y axis label.
# Take the x and y data and plot.
# use show() to see the output
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show

output_file('test.html')
p = figure(title = "test", x_axis_label='X-axis', y_axis_label='Y-axis')

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.circle(x, y, size=10, color='blue')
show(p)
```

```python
#2.

# In Bokeh, glyphs are visual markers or symbols used to represent data on a plot.
# They can include shapes, lines, markers, bars, and other graphical elements that help convey information visually.
# Glyphs are added to a Bokeh plot using specific glyph methods provided by the Bokeh library.
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show
output_file('Glyph.html')
plot = figure(title='Glyph Example', x_axis_label='X-axis', y_axis_label='Y-axis')

x_circle = [1, 2, 3, 4, 5]
y_circle = [6, 7, 2, 4, 5]
plot.circle(x_circle, y_circle, size=10, color='blue')

x_line = [1, 2, 3, 4, 5]
y_line = [5, 4, 2, 7, 6]
plot.line(x_line, y_line, line_width=2, color='red')

show(plot)
```

```python
#3.

# We can use following customisation:

# Title customization:
# title.text_color: Sets the color of the title text.
# title.text_font: Sets the font family of the title text.
# title.text_font_style: Sets the font style of the title text (e.g., 'italic', 'bold').
# title.text_font_size: Sets the font size of the title text.

# Axis label customization:
# xaxis.axis_label_text_color: Sets the color of the x-axis label text.
# xaxis.axis_label_text_font: Sets the font family of the x-axis label text.
# xaxis.axis_label_text_font_style: Sets the font style of the x-axis label text.
# xaxis.axis_label_text_font_size: Sets the font size of the x-axis label text.
# Similar properties are available for the y-axis (yaxis).

# Legend customization:
# legend.location: Sets the location of the legend (e.g., 'top_right', 'bottom_left', 'center').
# legend.title: Sets the title of the legend.
# legend.label_text_font: Sets the font family of the legend labels.
# legend.label_text_font_style: Sets the font style of the legend labels.
# legend.label_text_font_size: Sets the font size of the legend labels.
```
