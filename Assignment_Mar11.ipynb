```python
#1.

# The main difference between a t-test and a z-test lies in the assumptions and when they are applicable.

# A t-test is used when the population standard deviation is unknown or the sample size is small (typically less than 30).
# It assumes a normal distribution or a large enough sample for the central limit theorem to apply.
# The test statistic uses the sample standard deviation and has degrees of freedom based on the sample size.

# A z-test is used when the population standard deviation is known or the sample size is large (typically greater than 30).
# It also assumes a normal distribution or a large enough sample.
# The test statistic uses the known population standard deviation and critical values from the standard normal distribution.

# Example scenario for a t-test:
# comparing average scores of two groups with unknown population standard deviation and small sample sizes.

# Example scenario for a z-test: 
# comparing average height of a sample to the general population with a known standard deviation and large sample size.

# The choice depends on the availability of the population standard deviation and the sample size.
```

```python
#2.

# One-tailed test:
# It is used when the hypothesis predicts the relationship or difference in a specific direction.
# The critical region is defined on one side of the distribution, either the upper or lower tail.
# It is appropriate when there is prior knowledge or a specific hypothesis about the expected direction of the effect being tested.

# Two-tailed test:
# It is used when the hypothesis predicts a relationship or difference, but without specifying the direction.
# The critical region is divided into two equal parts, one in each tail.
# It is appropriate when there is no prior expectation or hypothesis about the direction of the effect, and you want to determine if there is a significant difference in either direction.
```

```python
#3.

# When we reject the null hypothesis when in reality is true, it is called Type 1 error.
# A pharmaceutical company conducts a clinical trial to test a new drug's effectiveness.
# They conclude that the drug is effective (reject the null hypothesis) when, in fact, it has no real effect.

# When we retain the null hypothesis when in reality it is false, it is called Type 2 error.
# Continuing with the pharmaceutical study example, suppose the null hypothesis (H₀) states that the drug has no effect on blood pressure, while the alternative hypothesis (H₁) suggests that the drug does lower blood pressure.
# If the researchers fail to reject the null hypothesis and conclude that the drug has no effect on blood pressure, even though it does lower blood pressure, it would be a Type 2 error.
```

```python
#4.

# Bayes's theorem is a fundamental concept in probability theory that allows us to update our beliefs or assess the probability of an event based on new evidence or information.
# It is named after Thomas Bayes, an 18th-century statistician. The theorem can be stated as:

# P(A|B) = (P(B|A) * P(A)) / P(B)

# Where,
# P(A|B) is the probability of event A given event B has occurred.
# P(B|A) is the probability of event B given event A has occurred.
# P(A) is the prior probability of event A.
# P(B) is the prior probability of event B.
# Here's an example to illustrate Bayes's theorem:

# Suppose there is a rare disease that affects 1 in 10,000 people. You go for a test that is 99% accurate, meaning it correctly identifies the disease 99% of the time for both healthy and infected individuals. Now, if you test positive for the disease, you might wonder what the probability is that you actually have it.

# Let's use the following notation:
# A: Having the disease
# B: Testing positive for the disease

# We are interested in calculating P(A|B), the probability of having the disease given that you tested positive.

# Given,
# P(A) = 1/10,000 (prior probability of having the disease)
# P(B|A) = 0.99 (probability of testing positive given that you have the disease)
# P(B|not A) = 0.01 (probability of testing positive given that you don't have the disease)
# Using Bayes's theorem:

# P(A|B) = (P(B|A) * P(A)) / P(B)
# = (0.99 * 1/10,000) / P(B)
```

```python
#5.

# Confidence Interval:
# Range of values used to estimate the unknown population parameters is called interval estimate.
# Interval estimates of population parameters are called confidence interval.

# We can calculate Confiedence interval using Z test, by following formula:
# C.I. = Point Estimate +/- Margin of Error
# C.I. = Sample_Mean +/- Z(alpha/2)*sample_std/square_root(sample_size)

# Let's say we want to estimate the average weight of a certain animal species.
# We collect a sample of 50 individuals and find that the sample mean weight is 10 kilograms, with a sample standard deviation of 1 kilogram.
# To calculate a 95% confidence interval, you can use the formula:

# Confidence Interval = Sample Mean ± (Z-value * Standard Error)

# With a 95% confidence level, the Z-value is approximately 1.96.
# The standard error is calculated as the sample standard deviation divided by the square root of the sample size.

# Let's assume the standard error is 0.1414 kilograms. Plugging in the values, we get:

# Confidence Interval = 10 ± (1.96 * 0.1414) = (9.719, 10.281)

# Therefore, with 95% confidence, we estimate that the true average weight of the animal species lies between 9.719 kilograms and 10.281 kilograms based on the sample data.
```

```python
#6.

# Suppose a certain disease affects 1% of the population.
# A diagnostic test for the disease has an accuracy rate of 95%.
# If a randomly selected individual tests positive for the disease, what is the probability that they actually have the disease?

# Solution:
# Let's use the following notation:
# A: Having the disease
# B: Testing positive for the disease

# We need to calculate P(A|B), the probability of having the disease given a positive test result.

# Given:
# P(A) = 0.01 (prior probability of having the disease)
# P(B|A) = 0.95 (probability of testing positive given that the individual has the disease)
# P(B|not A) = 0.05 (probability of testing positive given that the individual doesn't have the disease)

# Using Bayes' Theorem:
# P(A|B) = (P(B|A) * P(A)) / P(B)

# Substituting the values:
# P(A|B) = (0.95 * 0.01) / P(B)

# To calculate P(B), we need to consider the probability of testing positive in two scenarios:
# 1. Testing positive when having the disease (true positive): P(B|A) * P(A)
# 2. Testing positive when not having the disease (false positive): P(B|not A) * P(not A)

# P(B) = (0.95 * 0.01) + (0.05 * 0.99)

# Now, substitute the values back into the equation:

# P(A|B) = (0.95 * 0.01) / [(0.95 * 0.01) + (0.05 * 0.99)]
```

```python
#8.

# The margin of error in a confidence interval is a measure of the maximum likely difference between the sample estimate and the true population parameter.
# It indicates the level of uncertainty associated with the estimate.

# The margin of error is affected by several factors, including the desired level of confidence and the variability of the data.
# However, one significant factor is the sample size. Generally, a larger sample size tends to result in a smaller margin of error.

# When the sample size increases, the sample estimate becomes more representative of the population, leading to a more precise estimate.
# With a larger sample, the variability of the data is better captured, reducing the uncertainty and thus shrinking the margin of error.

# For example, consider estimating the average height of students in a school.
# With a small sample size of 20 students, the margin of error might be ±2 inches.
# However, if the sample size is increased to 200 students, the margin of error could decrease to ±0.5 inches.
# The larger sample provides a more reliable estimate with a narrower range of plausible values.
```

```python
#10.

# Null Hypothesis (H₀): The weight loss drug is not significantly effective. The population mean weight loss is equal to or less than zero.
# Alternative Hypothesis (H₁): The weight loss drug is significantly effective. The population mean weight loss is greater than zero.

# Given:
# Sample Size (n) = 50
# Sample Mean (x̄) = 6 pounds
# Sample Standard Deviation (s) = 2.5 pounds

# We will use a one-sample t-test since we are comparing the sample mean to a population mean.

# Step 1: Set the significance level (α). In this case, it is 0.05 (corresponding to a 95% confidence level).

# Step 2: Calculate the t-value using the formula:
# t = (x̄ - μ₀) / (s / √n)

# Where:
# μ₀ is the hypothesized population mean (in this case, μ₀ = 0).

# Plugging in the values:
# t = (6 - 0) / (2.5 / √50) ≈ 17.89

# Step 3: Determine the critical t-value from the t-distribution table or use statistical software. For a one-tailed test at a 95% confidence level with 49 degrees of freedom (n-1), the critical t-value is approximately 1.676.

# Step 4: Compare the calculated t-value with the critical t-value. If the calculated t-value is greater than the critical t-value, we reject the null hypothesis; otherwise, we fail to reject the null hypothesis.

# In this case, 17.89 > 1.676, so we reject the null hypothesis.
```
