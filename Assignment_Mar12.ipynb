```python
#1.

# c.i. = 0.95
# sample_mean = 50
# sample_std = 5

# C.I. = Point Estimate +/- Margin of Error
# C.I. = sample_mean +/- Z(alpha/2)*sample_std/square_root(sample_size)

# so confidence level lower bound and upper bound will be:

# lower_bound = 50 - 1.96*5/square_root(sample_size)
# upper_bound = 50 + 1.96*5/square_root(sample_size)

# lets assume sample_size = 100

# lower_bound = 50 - 1.96*5/square_root(100) = 49.02
# upper_bound = 50 + 1.96*5/square_root(100) = 50.98
```

```python
#2.

import scipy.stats as stat
import numpy as np

observed_data = [20, 15, 10, 15, 15, 25]
expected_data = [20, 20, 20, 10, 10, 20] * (int) (sum(observed_data)/100)

chisquare_test_statistics, p_value = stat.chisquare(observed_data, expected_data)

print('Chi Square Test :', chisquare_test_statistics)
print('P Value :', p_value)

significance_value = 0.05
dof = len(observed_data)-1
critical_value = stat.chi2.ppf(1-significance_value, dof)

print('Critical Value :', critical_value)

if chisquare_test_statistics > critical_value:
    print('Reject the null Hypothesis')
else:
    print('Accept the null Hypothesis')
```

```python
#4.

import statsmodels.stats.proportion as prop

n = 500
x = 60

lower_bound, upper_bound = prop.proportion_confint(x, n, alpha=0.05, method='normal')
print(f"95% Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")
```

```python
#5.

import numpy as np
from scipy.stats import norm

mean = 75
std_dev = 12
sample_size = 100

standard_error = std_dev / np.sqrt(sample_size)

confidence_level = 0.90
z_score = norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * standard_error
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"90% Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")
```

```python
#6.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10
x = np.linspace(0, 30, 500)

pdf = chi2.pdf(x, df)

plt.plot(x, pdf, label=f"df = {df}")
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density Function (PDF)")
plt.title(f"Chi-square Distribution (df = {df})")

mask = x <= 15
plt.fill_between(x[mask], pdf[mask], color='gray', alpha=0.3, label="Chi-square Statistic of 15")
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import t

sample_mean = 72
sample_std_dev = 10
sample_size = 30

null_hypothesis_mean = 70

standard_error = sample_std_dev / np.sqrt(sample_size)

t_statistic = (sample_mean - null_hypothesis_mean) / standard_error

alpha = 0.05

degrees_of_freedom = sample_size - 1

critical_t = t.ppf(1 - alpha / 2, degrees_of_freedom)

p_value = 2 * (1 - t.cdf(abs(t_statistic), degrees_of_freedom))

print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

if abs(t_statistic) > critical_t or p_value < alpha:
    print("The population mean is significantly different from 70.")
else:
    print("The population mean is not significantly different from 70.")
```
