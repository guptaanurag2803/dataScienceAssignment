```python
#1.

# Matplotlib is a python library used for visualisation.
# It contains a number tools used for creating graphs, charts and visualisation.

# It is used to explore, analyse and visualize data.
# It is used in different fields like science, finance, engineering, etc.

# Line plot, bar plot, scatter plot, histogram plot, box plot are some plots in pyplot. 
```

```python
#2.

# Scatter plot is a type of graph which shows the relationship between two variables.
# In scatter plot, each data point is represented by a dot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#3.

# subplot() function is to draw multiple plots in a single figure.
# This function creates a grid of subplots in a figure.
# The subplots function takes three arguments, which are the number of rows, number of columns, and the index of the subplot to be created.
# The index is a number that ranges from 1 to the total number of subplots that can be created in the grid.

import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.show()
```

```python
#4.

# Bar plot is a graphical representation of data.
# In bar plot, data is represented in the form of bars of different heights or lengths.
# It is used to show the changes in a variable over time or to compare the difference in values of domains or categories.

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.show()
plt.barh(company, profit)
plt.show()
```

```python
#5.

# A Box Plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.
# In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.show()

plt.boxplot(box2)
plt.show()
```
