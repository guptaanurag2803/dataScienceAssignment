# Google PlayStore

```python
#9.

import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

df.head()
rows, columns = df.shape

print('Rows:', rows)
print('Columns:', columns)
```

```python
#10.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 7))
plt.xticks(rotation=90)
sns.boxplot(x = df['Category'], y = df['Rating'])
```

```python
#11.

df.isnull().sum()

# from above code we can see that there are a lot of missing values in rating and also have some missing values in type, content rating, current ver and android ver also.

# Analyzing missing values is essential because they can impact your analysis in several ways:

# Missing Data Bias:
# If missing values are not handled properly, they can introduce bias into your analysis.
# The missing values may not be random and could be related to certain patterns or characteristics within the data, leading to biased conclusions.

# Incomplete Analysis:
# Missing values can result in incomplete analysis as they may prevent you from considering certain features or observations in your analysis, potentially leading to incomplete or inaccurate insights.

# Data Imputation:
# If you choose to impute or fill in missing values, it can impact the statistical properties of the dataset, such as means, medians, or correlations, which can further affect the validity of your analysis.

# Data Quality and Interpretation:
# Missing values can raise concerns about the overall data quality and reliability. 
# It is important to consider the extent of missing data and its potential impact on the interpretation of your analysis results.
```

```python
#12.

plt.figure(figsize=(13, 7))
plt.scatter(x = df['Size'], y = df['Rating'],marker = '+')
plt.title('App Size vs. Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.xticks([1, 100, 200, 300, 400], rotation=45)
plt.show()

# By visualizing the relationship between app size and rating using a scatter plot, you can observe if there is any pattern or correlation between the two variables.
# Points closer together may indicate a stronger relationship, while points spread out may indicate a weaker or no relationship.
```

```python
#13.

df['Price'] = df['Price'].str.replace('$', '')
df = df.drop(df[df['Price'] == 'Everyone'].index)
df['Price'] = df['Price'].astype(float)
avg_prices = df.groupby('Type')['Price'].mean()

plt.figure(figsize=(10, 6))
plt.bar(avg_prices.index, avg_prices.values)
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

```python
#14.

df['Installs'] = (df['Installs'].str).replace(',', '')
df['Installs'] = (df['Installs'].str).replace('+', '')
df['Installs'] = df['Installs'].astype(float)
top_10_apps = df.nlargest(10, 'Installs')

# Create a frequency table for the top 10 apps
frequency_table = top_10_apps['App'].value_counts()

print('Top 10:\n', top_10_apps[['App', 'Installs']], '\n\n')
print('Frequency Table:\n', frequency_table)
```
